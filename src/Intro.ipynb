```python
numbers = [25, 45, 80, 100, 120]
duplicados = []
pares= []
impares= []

for number in numbers:
    duplicados.append(number * 2)
    if number % 2 == 0:
        pares.append(number)
    else:
         impares.append(number)

pares
```
